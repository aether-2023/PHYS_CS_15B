```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
from scipy.optimize import curve_fit
```

```python
while 1>0:
    tName = input('input the trial name (M-L-A): ')
    if tName=='-1':
        break
    periods = np.array([])
    i=0
    while i<6:
        i+=1
        theP = int(input('data: '))/(500.0)
        periods=np.append(periods, theP)
    print(periods)
    mean = np.mean(periods)
    std = np.std(periods,ddof=1)
    aData=open('newDataHere/'+'40-'+tName+'-9','wb')
    pickle.dump([mean,std,periods],aData)
    aData.close()
```

```python
def func2(x,a):
    return a*x
```

```python
def func(x,g):
    return (2*np.pi)*((x/abs(g*100))**(0.5))
```

```python
def func1(x,a):
    return a+0*x
```

```python
def paPlot():
    xdata = np.linspace(0, 24, 50)
    a=24
    theMean = []
    theAmplitude = []
    theError = []
    while a>3:
        a-=3
        fileName = 'newDataHere/'+'40-40-'+str(a)
        aData = open(fileName, 'rb')
        meanP,stdP,P= pickle.load(aData)
        aData.close()
        theMean.append(meanP)
        theAmplitude.append(a)
        theError.append(stdP)
    popt, pcov = curve_fit(func1, theAmplitude, theMean)
    plt.figure()
    plt.plot(theAmplitude,theMean,'*')
    plt.errorbar(theAmplitude, theMean, yerr = theError, fmt='+')
    plt.plot(xdata, func1(xdata, *popt),
             label='fit: a=%5.3f' % tuple(popt))
    plt.xlabel('Amplitude (degree)')
    plt.ylabel('Period (seconds)')
    plt.legend()
    plt.show()
paPlot()
```

```python
def pmPlot():
    xdata = np.linspace(10, 90, 50)
    m=90
    theMean = []
    theMass = []
    theError = []
    while m>20:
        m-=10
        fileName = 'newDataHere/'+str(m)+'-40-9'
        aData = open(fileName, 'rb')
        meanP,stdP,P= pickle.load(aData)
        aData.close()
        theMean.append(meanP)
        theMass.append(m)
        theError.append(stdP)
    popt, pcov = curve_fit(func1, theMass, theMean)
    plt.figure()
    plt.plot(theMass,theMean,'*')
    plt.errorbar(theMass, theMean, yerr = theError, fmt='+')
    plt.plot(xdata, func1(xdata, *popt),
             label='fit: a=%5.3f' % tuple(popt))
    plt.xlabel('Mass (number of coins)')
    plt.ylabel('Period (seconds)')
    plt.legend()
    plt.show()
pmPlot()
```

```python
def plPlot():
    xdata = np.linspace(0, 80, 100)
    l=80
    theMean = []
    theLength = []
    theError = []
    while l>10:
        l-=10
        fileName = 'newDataHere/'+'40-'+str(l)+'-9'
        aData = open(fileName, 'rb')
        meanP,stdP,P= pickle.load(aData)
        aData.close()
        theMean.append(meanP)
        theLength.append(l)
        theError.append(stdP+0.05)
    popt, pcov = curve_fit(func, theLength, theMean)
    plt.figure()
    plt.plot(theLength,theMean,'*')
    plt.errorbar(theLength, theMean, yerr = theError, xerr=0.01, fmt='.',ecolor='b',color='g',elinewidth=2,capsize=4)
    plt.plot(xdata, func(xdata, *popt),
             color = 'r')
    plt.xlabel('Length (centimeter)')
    plt.ylabel('Period (seconds)')
    plt.show()
    print('g=%5.9f' % tuple(popt))
plPlot()
```

```python
def psqlPlot():
    xdata = np.linspace(0, 9, 50)
    l=80
    theMean = []
    theLength = []
    theError = []
    while l>10:
        l-=10
        fileName = 'newDataHere/'+'40-'+str(l)+'-9'
        aData = open(fileName, 'rb')
        meanP,stdP,P= pickle.load(aData)
        aData.close()
        theMean.append(meanP)
        theLength.append(l**(0.5))
        theError.append(stdP+0.05)
    popt, pcov = curve_fit(func2, theLength, theMean)
    plt.figure()
    plt.plot(theLength,theMean,'*')
    plt.errorbar(theLength, theMean, yerr = theError, xerr=0.01, fmt='+', label='sigma')
    plt.plot(xdata, func2(xdata, *popt),
             label='fit: a=%5.3f' % tuple(popt))
    plt.xlabel('Squareroot of Length (meter)')
    plt.ylabel('Period (seconds)')
    plt.legend()
    plt.show()
psqlPlot()
```
